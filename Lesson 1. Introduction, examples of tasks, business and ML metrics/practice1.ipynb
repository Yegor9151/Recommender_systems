### 1. Сравните метрики hit_rate@k, precision@k.<br>

##### Ответ:

- Hit rate@k - показывает факт покупки рекомендованного товара в формате да/нет. Без учета количества товаров. Можно проинтерпретировать, как - "работает ли можель вообще"

- Precision@k - показывает процентное соотношение купленных товаров от рекомендованных - является более точной метрикой в отличает от Hit rate@k. Можно проинтерпретировать, как - "насколько хорошо работает модель"

### 2. Какую матрику использовать предпочтительно и почему.

##### Ответ:

Думаю что предпочтительнее использовать Precision@k, так как она дает более подробную информацию о качестве системы

### 3. Привидите пример 2-3 задач (опишите, что является клиентом, что товаром), в которой более уместно использовать метрику hit_rate?

##### Ответ:

Судя по тому, что hit_rate учитывает сам факт покупки из рекомендуемых можно предположить, что такая метрика будет уместна в тех случаях когда рекомендуемый товар имеет большую цену либо выбор заведомо только 1. 

- Например если рекомендуется бытовая техника, тут клиентом будет посетитель интернет ресурса, а товаром техника.
- Или отель для отдыха, или номер отеля, тут клиент так же человек а товар номер или отель
- Если организация ищет специалиста на ставку, тут будет клиентом работодатель или кадровый агент, а товаром человек ищущий работу

### 5. Какие еще матрики (Вы можете вспомнить уже пройденные Вами или посмотреть в интернете) могут использоваться для рекомендательных систем (привидите примеры метрики и чем являются интеракции, чтобы эта матрика могла быть использована).

##### Ответ:

- __f1 score__ - так как она позволяет увидель гармоническую оценку пресказаных результатов. Например если подразумевается, что человек должен выбрать несколько позицый, то можно оценить насколько точно совпадают предложенные вариенты с тем, что он сам выбрал.
- __confusion_matrix__ - позволит нам увидеть более полную картину типа: _рекомендовано и куплено / рекомендовано, но не куплено / не рекомендовано, но куплено / не рекомендовано и не куплено_
