### В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной фильтрации?

##### Ответ:
- В __гибридных системах__ подается не только __матрица интеракций__, но и __внешние фичи__ (LightFM)
- __коллаборативная фильстрация__ - на вход подается только __матрица интеракций__ (ALS)

### Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы

##### Ответ:
- Если нужно учитывать не только факт взаимодействие, но и ресурс на котором пользователь взаимодействовал с товаром, например если один и тот же товар был куплен на разных сайтах, или переход на сайт был с других источников. Тогда нужно будет учитывать и эти показатели, так как у пользователей разных ресурсов могут различаться потребности.
- Если нужно будет учитывать время или место взаимодействия для прогнозирования рекоммендаций. У пользователей могут меняться предпочтения в зависимости от этих показателей.

3) С какими моделями сравнивают LightFM? Опишите их основные идеи кратко

- __MF__ - матричная модель User х Item с функцией сигмойды
- __LSI-LR__ - включает в себя логистическую регрессию. Основывется на факторизации матрица содержания. 
- __LSI-UP__ - гибридная модель, которая представляет профили пользователей (UP), как линейную комбинацию предметов по векторам, затем применяет LSI к получивщейся матрице, чтобы получить скрытые представления пользователей и элементов.
